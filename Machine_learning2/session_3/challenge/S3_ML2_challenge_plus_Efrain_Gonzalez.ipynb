```python
# NOTA:  Para este ejercicio utilizaremos una ventana de 7 dias y un baremo de 2 veces la desviacion estandar para 
# definir los rangos donde debe encontrarse el valor de la accion y considerarlo 'normal'
```

```python
# NOTA: Al revisar el dataset se observan algunos valores nulos, los cuales afectan al calculo de la ventana 
# deslizante.

# Optamos de cara a seguir con el analisis aplicar interpolaciones pero al tratarse de unos valores con mucha
# volatibilidad, este el criterio de tratamiento de nulos debe revisarse con mucho cuidado
```

```python
df['Close'].isna().any(axis=0)
```

```python
df_close=df['Close'].interpolate(method='linear', axis=0).bfill().ffill()
```

```python
# Definimos una funcion wrapper que realiza los calculos de la ventana deslizante y baremos, adicionalmente
# contrasta el valor de la accion con dichos baremos para determinar si se trata de una anomalia techo o suelo.
# Como resultado final devuelve el dataset con todos los calculos.

def anomaly_review(df, ticker, wind, sigma):

    result = pd.DataFrame()
    result["Close"] = df[ticker]
    result["suelo"] = df[ticker].rolling(window=wind)\
        .mean() - (sigma * df[ticker].rolling(window=wind).std())
    result["techo"] = df[ticker].rolling(window=wind)\
        .mean() + (sigma * df[ticker].rolling(window=wind).std())
    result["anom_techo_flag"] = result.apply(
        lambda row: 1 if (row[0]>=row["techo"]) else 0, axis=1)
    result["anom_suelo_flag"] = result.apply(
        lambda row: 1 if (row[0]<=row["suelo"]) else 0, axis=1)
    result["anom_techo"] = result.apply(
        lambda row: row[0] if (row[0]>=row["techo"]) else 0, axis=1)
    result["anom_suelo"] = result.apply(
        lambda row: row[0] if (row[0]<=row["suelo"]) else 0, axis=1)
    result["ticker"] = ticker
    return result
```

```python
wind = 7
sigma = 2
tickers = df_close.columns
anoms = pd.DataFrame()
for ticker in tickers:
    review = anomaly_review(df_close,ticker,wind,sigma)
    review[["Close","suelo","techo","anom_techo","anom_suelo"]].plot(title=ticker)
    anoms = anoms.append(review, ignore_index=False)
anoms = anoms.drop(columns=['suelo', 'techo'])
```

```python
# Creamos una tabla pivotada para facilitar el analisis de los datos
anoms_pivot=pd.pivot_table(anoms,
                        index='Date',
                        columns='ticker',
                        values=['anom_techo_flag','anom_suelo_flag'],
                        fill_value=0,
                       )
anoms_pivot.head()
```

 * Con una inversión de 1000€ durante el año ¿cuanto dinero habríamos ganado o perdido utilizando la estretegia de inversión considerada?

```python
anomalies = tmp[(tmp['anom_techo_flag']==1) | (tmp['anom_suelo_flag']==1)][['Close','anom_techo_flag','anom_suelo_flag','ticker']].copy()
anomalies.sort_index()
```

```python
# Con la inversion inicial se recorreran las anomalias y en base al tipo de anomalia se realizara 
# una inversion o una venta de acciones.

# Como criterio conservador y evitar problemas por multiples compras o ventas consecutivas, y para no quedarnos
# sin capital y perder oportunidades, en cada transaccion solo se invertira una fraccion de lo disponible

capital = 1000
# Este parametro define 'cuanto' de lo que se tiene (accion o capital) se usa en la transaccion
ratio_por_transaccion = 0.04  
# Inicializamos la cartera de acciones
inversiones = dict((x,0) for x in assets)
print('Empezamos con estas cantidad de acciones:')
display(inversiones)

print('Este es el log de compras/ventas')

for index, row in anomalies.iterrows():
    if row['anom_techo_flag'] == 1:
         # Debemos vender las acciones que tenemos (por ser conservador solo venderemos 50% de las acciones)
         ticker = row['ticker']
         acciones_a_vender = inversiones[ticker] * ratio_por_transaccion
         inversiones[ticker] = inversiones[ticker] - acciones_a_vender
         capital = capital + acciones_a_vender*row['Close']
         print('Dia {0}, Venta, {1}, acciones_vendidas {2}, precio {3}, capital {4}'.format(index, 
                                                                                   ticker,
                                                                                   acciones_a_vender,
                                                                                   row['Close'],
                                                                                   capital
                                                                                   ))
    if row['anom_suelo_flag'] == 1:
         # Debemos comprar (por ser conservador solo usaremos un 50% del capital que se disponga)
         ticker = row['ticker']
         capital_a_invertir = capital * ratio_por_transaccion
         capital = capital - capital_a_invertir
         acciones_a_comprar = capital_a_invertir / row['Close']
         inversiones[ticker] = inversiones[ticker] + acciones_a_comprar
         print('Dia {0}: Compra, {1}, acciones_compradas {2}, precio {3}, capital {4}'.format(index, 
                                                                                              ticker,
                                                                                              acciones_a_comprar,
                                                                                              row['Close'],
                                                                                              capital
                                                                                             ))

print('Al final del backtesting nos quedamos con las siguietnes acciones sin vender:')
display(inversiones)

print('El capital residual luego de las operaciones fue:')
display(capital)
```

```python
# No obstante para finalizar el calculo se deben vender las acciones residuales al precio del ultimo dia 
# que tenemos en el dataset para cuantifificar el capital total

# Los precios de las acciones al final del dataset son
df_close.iloc[-1]
```

```python
for ticker, acciones in inversiones.items():
    venta = acciones * df_close.iloc[-1][ticker]
    capital = capital + venta
    inversiones[ticker]=0
print('Al final del ejercicio nos quedamos con el siguiente capital')
capital
```

 * ¿Qué diferencias obtienes entre la aproximación sencilla en la detección de anomalías y Isolation forest?


```python
# Procedemos a crear una funcion wrapper que nos ayude a elaborar el modelo y aplicarlo para
# cada ticker.

from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import IsolationForest

def anomaly_review_IsolationForest(df, ticker, outliers_fraction):
    scaler = StandardScaler()
    np_scaled = scaler.fit_transform(df[ticker].values.reshape(-1,1))
    data = pd.DataFrame(np_scaled)
    model =  IsolationForest(contamination=outliers_fraction)
    model.fit(data) 
    result = pd.Series(model.predict(data))
    return result
```

```python
# Iteramos para cada ticker y nos devuelve las graficas con sus anomalias.
anoms_IsolationForest = pd.DataFrame()
for ticker in tickers:
    review = pd.DataFrame()
    review['Close'] = df_close[ticker]
    review.loc[:,'ticker'] = ticker
    review['anom_flag'] = anomaly_review_IsolationForest(df_close,ticker, 0.02).values
    review['anom_flag'] = review['anom_flag'].apply(lambda x: 0 if (x == 1) else 1)
    review['anom_value'] = review['anom_flag']*review['Close']
    review[["Close","anom_value"]].plot(title=ticker)
    anoms_IsolationForest = anoms_IsolationForest.append(review, ignore_index=False)
```

```python
# El modelo solo indica que hubo una anomalia pero, a partir de esa informacion
# debemos ver si se trata de un techo o un suelo para aplicar la politica de compra/venta

# Como primer aproximacion usaremos el valor anterior a la anomalia para determinar si es un
# techo o suelo.

anoms_IsolationForest
```

```python

anoms_IsolationForest['close_shifted_1d'] = anoms_IsolationForest.groupby('ticker', group_keys=False)['Close'].apply(lambda x: x.shift(1)) 
anoms_IsolationForest = anoms_IsolationForest.dropna()
anoms_IsolationForest['close_differences_1d'] = anoms_IsolationForest['Close'] - anoms_IsolationForest['close_shifted_1d']
anoms_IsolationForest['anom_techo_flag']=anoms_IsolationForest.apply(lambda row: 1 if ((row['anom_flag']==1) and (row['close_differences_1d']>0)) else 0, axis=1) 
anoms_IsolationForest['anom_suelo_flag']=anoms_IsolationForest.apply(lambda row: 1 if ((row['anom_flag']==1) and (row['close_differences_1d']<0)) else 0, axis=1)
anoms_IsolationForest_to_review = anoms_IsolationForest[(anoms_IsolationForest['anom_flag'] == 1)].copy()
anoms_IsolationForest_to_review
```
