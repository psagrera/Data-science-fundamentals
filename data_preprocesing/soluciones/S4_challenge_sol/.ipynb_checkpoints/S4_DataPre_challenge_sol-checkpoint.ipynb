```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression

import warnings
warnings.filterwarnings('ignore')

import random
random.seed(1234)

tr = pd.read_csv("data/transmission_small.csv")
```

`Se necesita entender la magnitud del problema. Informe del posible desbalanceo de los datos, tanto por escrito como mediante la gráfica que usted considere oportuna`

```python
tr['Class'].value_counts(normalize=True)
```

```python
tr['Class'].value_counts().plot(kind='bar', title='Class')
```

`También deberá conocer en qué rango se manejan los datos y con qué estamos trabajando. Una gráfica de puntos podría ser lo más adecuado ahora.`

```python
colors = tr['Class'].map({0 : "#1f77b4", 1: "#ff7f0e"})
plt.scatter(x = tr['V2'], y = tr['V1'], c=colors)
```

`Le proponemos ahora una función sencilla para evaluar su sistema de mejora de los datos. Esta función no deberá modificarse y como único requisito deberá de suministrar los datos como parámetro.`

```python
def evaluar_sistema(data_variables, class_variable):
    # Recibe:
    # - data_variables: dataframe con dos columnas siendo estas las variables de datos.
    # - class_variable: dataframe con una columna siendo esta la clase de transmisión.
    # Imprime:
    # - puntuación de clasificación.
    print("Evaluando...")
    clf = LogisticRegression().fit(data_variables, class_variable.values.ravel())
    predicted = clf.predict(data_variables)
    
    puntos = 0
    for ci, pi in zip(class_variable.values.ravel(), predicted):
        if ci != pi:
            n = 1 if ci == 0 else 50
        else:
            n = 0
        puntos += n
        
    print("Puntuación obtenida: {} (más alta es peor)".format(puntos))
```

```python
evaluar_sistema(tr[['V1', 'V2']], tr[['Class']])
```

```python
## Introduce todos los pasos que consideres necesarios y posteriormente, evalúa tus datos. ¡Que el lado oscuro te acompañe!
```

```python
from imblearn.under_sampling import RandomUnderSampler

rus = RandomUnderSampler(random_state = 0)
rus.fit_sample(tr[['V2', 'V1']], tr[['Class']].values.ravel())
print('Filas que nos quedamos:', len(rus.sample_indices_))
```

```python
tr_undersampled = tr.iloc[rus.sample_indices_]
tr_undersampled['Class'].value_counts().plot(kind='bar', title='Class')
```

```python
colors = tr_undersampled['Class'].map({0 : "#1f77b4", 1: "#ff7f0e"})
plt.scatter(x = tr_undersampled['V2'], y = tr_undersampled['V1'], c=colors)
```

```python
evaluar_sistema(tr_undersampled[['V1', 'V2']], tr_undersampled[['Class']])
```

```python
# RandomUndersampling parece obtener muy buenos resultados, 
# pero nos hemos cargado prácticamente toda la estructura del dataset....
```

```python
## KNN undersampling
```

```python
from imblearn.under_sampling import EditedNearestNeighbours

enn = EditedNearestNeighbours(sampling_strategy='auto', n_neighbors = 51)
enn.fit_sample(tr[['V2', 'V1']], tr[['Class']].values.ravel())
print('Filas que nos quedamos:', len(enn.sample_indices_))
```

```python
tr_enn = tr.iloc[enn.sample_indices_]
tr_enn['Class'].value_counts().plot(kind='bar', title='Class')
```

```python
colors = tr_enn['Class'].map({0 : "#1f77b4", 1: "#ff7f0e"})
plt.scatter(x = tr_enn['V2'], y = tr_enn['V1'], c=colors)
```

```python
evaluar_sistema(tr_enn[['V1', 'V2']], tr_enn[['Class']])
```

```python
## Vamos a combinar con Adasyn, parece hemos abierto una frontera parcial.
```

```python
from imblearn.over_sampling import ADASYN
ada = ADASYN()
X, y = ada.fit_sample(tr_enn[['V1', 'V2']], tr_enn[['Class']])

# convertimos a DataFrame por comodidad y renombramos las columnas de datos en X a los nombres que teníamos previamente.
alerts_ada = pd.DataFrame(X).rename(columns = {0 : 'confidence', 1 : 'distance'})
alerts_ada['Class'] = y

print("Datos después de ADASYN: {}".format(len(y)))
```

```python
colors = alerts_ada['Class'].map({0 : "#1f77b4", 1: "#ff7f0e"})
plt.scatter(x = alerts_ada['V2'], y = alerts_ada['V1'], c=colors)
```

```python
evaluar_sistema(alerts_ada[['V1', 'V2']], alerts_ada[['Class']])
```

```python
from imblearn.over_sampling import SMOTE
smote = SMOTE()
X, y = smote.fit_sample(tr_enn[['V1', 'V2']], tr_enn[['Class']])

# convertimos a DataFrame por comodidad y renombramos las columnas de datos en X a los nombres que teníamos previamente.
alerts_smote = pd.DataFrame(X).rename(columns = {0 : 'confidence', 1 : 'distance'})
alerts_smote['Class'] = y

print("Datos después de SMOTE: {}".format(len(y)))
```

```python
colors = alerts_smote['Class'].map({0 : "#1f77b4", 1: "#ff7f0e"})
plt.scatter(x = alerts_smote['V2'], y = alerts_smote['V1'], c=colors)
```

```python
evaluar_sistema(alerts_ada[['V1', 'V2']], alerts_ada[['Class']])
```
