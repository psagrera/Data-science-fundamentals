```python
MADRID_COORDS = [40.4167,  -3.70325]
m = folium.Map(MADRID_COORDS, zoom_start=13)
m
```

```python
for index, row in estaciones.iterrows():
    folium.CircleMarker([row['LATITUD'], row['LONGITUD']],
                        radius=5,
                        popup=row['Calle'],
                        fill_color="#3db7e4",
                       ).add_to(m)
m
```

```python
from faker import Faker
import random
random.seed(12)
faker = Faker() 

estaciones = pd.read_csv("data/bases_bicimad.csv", encoding = "latin-1", sep = ";")
estaciones.head()

print("estaciones ", estaciones.shape)

random_coordinates = [(39.4697500, -0.3773900), (37.2548200,-7.2043300), (40.070393, -2.137416), (37.984047, -1.128575),
                      (39.4677500, -0.3753900), (37.2544200,-7.2045300), (40.071393, -2.137416), (37.984147, -1.128675)]

for i in range(8):
    fake_city = random_coordinates[random.randint(0, 7)]
    fake_series = pd.Series(estaciones.iloc[random.randint(0, 170)])
    fake_series['Número'] = random.randint(0, 100)
    fake_series['LONGITUD'] = fake_city[1]
    fake_series['LATITUD'] = fake_city[0]
    estaciones = estaciones.append(fake_series)
    
print("estaciones ", estaciones.shape)
```

```python
from sklearn.utils import shuffle
estaciones = shuffle(estaciones)

estaciones.to_csv("data/estaciones_fake.csv", header = True, index = False)
estaciones = pd.read_csv("data/estaciones_fake.csv")
```
