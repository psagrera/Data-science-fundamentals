## Generación de números aleatorios

Las funciones de generación de números aleatorios nos permiten generan aleatoriedad con el ordenador. La función más simple para generar números aleatorios con `numpy` es `random_sample()` que nos devuelve un número entre `[0, 1]`:

```python
np.random.random_sample()
```

Esta función acepta como parámetro las dimensiones de una matriz o de un vector

```python
np.random.random_sample(10)
```

### Distribución uniforme

La función `random_sample` nos devuelve números que siguen una distribución uniforme, lo podremos comprobar dibujando un histograma de los números:

```python
pd.DataFrame(np.random.random_sample(10)).plot.hist()
```

Podemos ver que a medida que aumentamos el número de instancias aleatorias el histograma obtenido se parece más a una distribución uniforme

```python
pd.DataFrame(np.random.random_sample(100)).plot.hist()
```

```python
pd.DataFrame(np.random.random_sample(1000)).plot.hist()
```

```python
pd.DataFrame(np.random.random_sample(100000)).plot.hist()
```

### Distribución normal

Como podemos ver la distribución uniforme está asociada a sucesos equiprobables como por ejemplo el lanzamiento de un dato. No obstante muchos sucesos en la naturaleza no siguen esta distribución. Una distribución muy habitual es la distribución normal, que está asociada a procesos en los que existen valores centrales muy probables y los valores valores extremos son exponencialmente menos probables. 

Por ejemplo la altura media de los españoles es un ejemplo de normal. La mayor parte de los españoles están alrededor de 1,75 y tenemos exponencialmente menos casos a los lados.

Podemos construir distribuciones normales mediante la función `np.random.normal()`. Una distribución normal está caracterizada por 2 parámetros $\mu$ y $\sigma$.

* $\mu$: representa la media de los números a generar
* $\sigma$: representa la desviación estándar

Por ejemplo si queremos generar una distribución normal estándar ($\mu$=0, $\sigma$=1) lo podemos hacer de la siguiente manera

```python
mu = 0
sigma = 1
pd.DataFrame(np.random.normal(mu, sigma, 1000)).plot.hist()
```

Podemos jugar con los valores creando diferentes tipos de normales.

```python
mu = 10
sigma = 10
pd.DataFrame(np.random.normal(mu, sigma, 1000)).plot.hist()
```

### Distribución de pareto

La distribución de pareto también es una distribución que está asociado a muchos procesos reales en los que hay pocos elementos que tienen muchas ocurrencias y luego una larga cola de sucesos mucho menos probables. Situaciones que suelen seguir una distribución de pareto son:

* Ventas de discos de música
* Distribución de riqueza en las personas
* Siniestralidad en seguros

Para generar distribuciones de pareto podemos utilizar la función `np.random.pareto()` que recibe como argumentos la forma de la distribución y el número de elementos a generar.

```python
forma = 0.1
pd.DataFrame(np.random.pareto(forma, 100)).plot.hist()
```

```python
forma = 1
pd.DataFrame(np.random.pareto(forma, 100)).plot.hist()
```

```python
forma = 10
pd.DataFrame(np.random.pareto(forma, 100)).plot.hist()
```

## Simulación de fuentes de datos

Es muy común en escenarios de IoT simular dispositivos para poder realizar hipótesis sobre los datos a manejar o bien para poder evaluar la escalabilidad de un sistema.

Supongamos que tenemos un dispositivo que realiza mediciones de temperatura, posición GPS (latitud y longitud) y nivel de bateria y queremos generar 10.000 dispositivos virtuales que nos permitan entender como se puede comportar un sistema IoT.

Lo primero que vamos a hacer es declarar un dataframe vacío con las columnas a generar:

```python
df = pd.DataFrame(columns=['Batt', 'Temp','GPS_lat', 'GPS_long'])
```

Ahora por cada sensor, debemos decidir que distribución sigue y caracterizarla de manera que la simulación sea lo más realista posible.

Por ejemplo para modelizar el nivel de batería, asumiendo que la batería decrece de una forma lineal, podemos asumir que la distribución de los niveles de batería sigue una distribución uniforme, entre 0 = 0% y 1 = 100%:

```python
df['Batt'] = np.random.random_sample(10000)
df['Batt'].plot.hist()
```

El siguiente sensor que vamos a modelizar es la temperatura. Para ello asumiremos que la temperatura sigue una distribución normal, en este caso asumiremos que estamos en Madrid, y la temperatura media ($\mu$) es de 17ºC y la desviación típica ($\sigma$) es de 10ºC

```python
df['Temp'] = np.random.normal(17.0, 10.0, 10000)
df['Temp'].plot.hist()
```

Por último nos queda modelizar posiciones GPS. En este caso, asumiremos que el rango de los dispositivos están alrededor de un punto central de las coordenadas de Madrid. Si las coordenadas de Madrid

```python
Madrid_lat = 43.0
Madrid_long = 30.0
desviacion = 0.1

df['GPS_lat'] = np.random.normal(Madrid_lat, desviacion, 10000)
df['GPS_long'] = np.random.normal(Madrid_long, desviacion, 10000)
```

```python
df['GPS_lat'].plot.hist()
```

De esta forma hemos conseguido simular una visión estática de 10.000 dispositivos

```python
df.head()
```

## El método de montecarlo

El método de montecarlo (https://es.wikipedia.org/wiki/M%C3%A9todo_de_Montecarlo) nos permite analizar un sistema complejo mediante el análisis estadístico de casos de prueba aleatorios.

Es decir, supongamos que disponemos de una función (o un proceso físico, o cualquier proceso) de la que desconocemos como funciona por dentro, pero somos capaces de introducir variables de entrada y obtener su resultado. Si introducimos combinaciones aleatoria de los parámetros de entrada y analizamos los resultados podemos inferir muchas cosas como:

* El rango de los valores de salida
* La dependencia de los valores de entrada con la salida.

Por ejemplo, disponemos de un motor de combustión, que es un elemento físico muy complejo. Queremos saber su consumo de gasolina respecto a varias variables de entrada: temperatura del motor, tipo de aceite, revoluciones... Desde un punto de vista funcional el motor se comporta como la siguiente funcional, de la que obviamente no conocemos su contenido:

```python
def consumo_motor_gasolina(temperatura, tipo_aceite, revoluciones):
    # --- Este código no lo conocemos ---
    import random

    consumo = 5.0 # Consumo base
    if temperatura < 80 or temperatura > 100:
        consumo *= 1.1
    elif temperatura < 95 or temperatura > 95:
        consumo *= 1.05
    else:
        consumo *= 1
    
    if tipo_aceite == 0:
        consumo *= 0.99
    elif tipo_aceite == 2:
        consumo *= 1.04
    
    revoluciones = max(revoluciones, 1000.)
    revoluciones = min(revoluciones, 5000.)
    revoluciones_norm = (revoluciones - 1000.) / 4000.
    consumo *= 1. + (revoluciones_norm / 10.)
    
    consumo += random.uniform(-0.04, 0.04)
    # -----------------------------------
    return consumo
```

El método de montecarlo nos permite, en este ejemplo, analizar el consumo del motor introduciendo variaciones sobre las variables de entrada y analizando la salida.

Para ello vamos a crear un dataframe con las combinaciones de parámetros de entrada a estudiar.

```python
df_intentos = pd.DataFrame(columns=['temperatura', 'tipo_aceite','revoluciones'])
```

El siguiente paso es crear los valores de cada una de las variables.

Para la temperatura del motor asumiremos que sigue una distribución normal, con $\mu=90ºC$ y $\sigma=10ºC$

```python
df_intentos['temperatura'] = np.random.normal(90.0, 10.0, 10000)
df_intentos['temperatura'].plot.hist()
```

En el caso de los tipos de aceite, como es una variable discreta utilizaremos la función `random.choice` que nos permite indicar una lista con los posibles valores a generar, el número de items a generar y la matriz de probabilidad de cada elemento:

```python
tipos_aceite = [0, 1, 2] # Posibles valores (extra, normal, normal)
tipos_aceite_p = [0.1, 0.6, 0.3] # función de probabilidad
df_intentos['tipo_aceite'] = np.random.choice(tipos_aceite, 10000, p=tipos_aceite_p)
df_intentos['tipo_aceite'].value_counts()
```

Y por último las revoluciones. Al igual que las temperatura la modelaremos mediante una distribución normal.

```python
df_intentos['revoluciones'] = np.random.normal(3000.0, 1000.0, 10000)
df_intentos['revoluciones'].plot.hist()
```

Como podemos ver, la distribución normal genera valores inferiores a 0. Como estos valores no son correctos en nuestro dominio del problema, los valores negativos les sustituiremos por 0. 

```python
df_intentos[df_intentos.revoluciones < 0] = 0
df_intentos['revoluciones'].plot.hist()
```

En este momento tenemos en `df_intentos` un dataframe con 10.000 posibles variaciones de los parámetros.

```python
df_intentos.head()
```

Ahora lo que vamos a hacer es ejecutar la función desconocida por cada una de las posibles variaciones. En este caso haremos una función de envoltorio (wrapper) para poder ajustar la entrada de datos.

```python
def consumo_motor_gasolina_wrapper(row):
    return consumo_motor_gasolina(row['temperatura'], row['tipo_aceite'], row['revoluciones'])

df_intentos['output'] = df_intentos.apply(consumo_motor_gasolina_wrapper, axis=1)
```

Una vez ejecutada la función el primer paso es observar la variable de salida:

```python
df_intentos['output'].plot.hist()
```

Como podemos ver en la figura, la distribución de salida se asemeja a una distribución normal con un valor medio de 5.6. Es importante observar que a pesar de la variabilidad de los valores de entrada la salida no presenta una variabilidad muy grande, ya que en muchas ocasiones los valores se anulan unos a otros.

El siguiente paso es analizar la relación de cada variable con el valor de salida, para ello vamos a utilizar una función que se llama `scatter_matrix` que nos muestra las relaciones entre todas las variables de un dataframe:

```python
from pandas.plotting import scatter_matrix
scatter_matrix(df_intentos, alpha=0.2, figsize=(16, 16), diagonal='kde')
```
