## C1 -RELACIONES TEMPORALES

¿Cómo son las ventas las últimas 20 semanas?

Selecciona los gráficos me mejor expliquen la tendencia de ventas, tanto general como por alguna dimensión 

```python
fecha=pd.Grouper(key='order_Dte', freq='W')

last_20_weeks = df_datos['2016-09-01':'2017-01-31']

last_20_weeks

campos_filas=[fecha]
campos_valores=['Sales']
agregados={np.mean}
#
ventas=pd.pivot_table(last_20_weeks,
                                    index=campos_filas,
                                    values=campos_valores,
                                    aggfunc=agregados)



fig, ax = plt.subplots()
ventas.plot(kind='line',y='Sales',ax=ax,title='Sales last 20 weeks',color='green')
ax.grid(True)
plt.show()

```

```python
# Filtramos solo un segmento
filtro_segmento = last_20_weeks['Segment'] == "HomeOffice"

df_datos_tmp = last_20_weeks[filtro_segmento]
fecha=pd.Grouper(key='order_Dte', freq='W')
campos_filas=[fecha]
campos_valores=['Sales']
ventas_segmento=pd.pivot_table(df_datos_tmp,
                                    index=campos_filas,
                                    values=campos_valores,
                                    aggfunc=agregados)


ax = plt.gca()
ventas_segmento.plot(kind='line',y='Sales',ax=ax,title='Sales last 20 weeks HomeOffice',color='red')
ax.grid(True)
plt.show()
```

## C2 - PARTES DE UN TODO

¿Qué categoría de producto es la más vendida?   
¿En qué porcentaje?  
¿Cuales son los gráficos que mejor nos muestran esa información ?  



```python
mas_vendida = df_datos.groupby('Category').agg({'Sales':'mean'})
mas_vendida['percent'] = mas_vendida['Sales']*100/mas_vendida['Sales'].sum()
labels = 'Techology', 'Office Supplies', 'Furniture'
fig1, ax1 = plt.subplots()
ax1.pie(mas_vendida.percent,labels=labels,autopct='%1.1f%%',shadow=True, startangle=90)
ax1.axis('equal') 
```

## C3 - COMPARATIVAS

¿Cómo evoluciona cada categoría de producto en estos **dos** ultimos años?

```python
fecha=pd.Grouper(key='order_Dte', freq='W')

last_2_years = df_datos['2015-01-01':'2017-01-31']

df_pivot = last_2_years.pivot_table(values='Sales',index=fecha,columns='Category')
df_pivot = df_pivot.dropna()
ax = plt.gca()
df_pivot.plot(kind='line',ax=ax,figsize=(20, 10))
ax.grid(True)
```

```python
# Ventas por Catergoria,Subcategoria
ventas_cat = cat.groupby(['Category','Subcategory']).agg({'Sales':'mean'})#.plot(kind='barh',stacked=True,color='green')
ventas_cat.sort_values('Sales',ascending=True).plot(kind='barh',color='green')
```

```python
# Transporte de las mercancias 
transporte = df_datos.groupby('Mode').agg({'Sales':'mean'})
transporte['percent'] = transporte['Sales']*100/transporte['Sales'].sum()
labels = 'Regular Air', 'Delivery Truck', 'Express Air'
fig1, ax1 = plt.subplots()
ax1.pie(transporte.percent,labels=labels,autopct='%1.1f%%',shadow=True, startangle=90)
ax1.axis('equal') 
```
