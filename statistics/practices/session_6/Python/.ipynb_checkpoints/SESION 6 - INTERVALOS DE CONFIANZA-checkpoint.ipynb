
# Sobre intervalos de Confianza

# Estimador: *tipos*
Una estimacin consiste en apostar por un valor para un parmetro poblacional, habitualmente calculando tambin un rango de valores entre los que puede situarse dicho parmetro poblacional. Por tanto, un intervalo de confianza es simplemente un rango de valores en el que confiamos que se contenga el parmetro poblacional (Altman, 1991).
>Puede decirse que el anlisis estadstico inferencial conduce a dos clases de resultados: valores p e intervalos de confianza.

Las estimaciones pueden ser de dos tipos:
>**Estimacin puntual**: obtendremos un punto, un valor, como estimacin del parmetro desconocido, quedando sin especificar cmo de buena es la aproximacin.
>> Si a partir de una muestra se afirma que el 70% de una poblacin es fumadora se est haciendo una estimacin puntual de la proporcin de fumadores.

>**Estimacin por intervalo**: proporciona un rango de posibles valores, dentro del cual estimamos (bajo cierta probabilidad) que estar el parmetro desconocido.
>>Si a partir de una muestra se afirma que el porcentaje de la poblacin que es fumadora se sita entre un 23 y 52% se est haciendo una estimacin por intervalo de la proporcin poblacional de fumadores



En general, se puede afirmar que un estimador puede ser:
+ **Insesgado**: cuando el estimador muestral coincide con el poblacional, sin ninguna desviacin sistemtica. Es decir, un estimador es insesgado si su valor medio sobre todas las posibles muestras de tamao n coincide con el parmetro poblacional.
+ **Sesgado**: cuando el estimador muestral no se corresponde con el poblacional, debido a un error sistemtico.
+ **Eficiente**: es aquel estimador muestral de igual valor a otros, que la varianza muestral es mnima. De esta forma, se tendr una mayor confianza en que la estimacin resultante de la muestra finalmente seleccionada est prxima al parmetro poblacional.
+ **Suficiente**: aquel estimador que utiliza toda la informacin contenida en la muestra.
>>Un buen estimador es aquel que es insesgado, eficiente y suficiente
(Rothman y Greenland, 1998)



# Qu es un intervalo de confianza?

Las estimaciones puntuales obtenidas a partir de una muestra *diferirn del parmetro poblacional* y, en consecuencia, quedar un *margen de incertidumbre* que se expresa en trminos del error estndar del
estimador. 

Por consiguiente, podemos querer disponer de una medida del parmetro poblacional que incorpore tanto la estimacin puntual como su error estndar. Esta medida es el **intervalo de confianza (I.C.)**, *que facilita un rango de valores dentro del cual se encontrar el verdadero valor del parmetro poblacional con un cierto grado de confianza.*

# Factores de los que depende un intervalo de confianza
El clculo de un intervalo de confianza depende principalmente de los siguientes factores:
- Tamao de la muestra seleccionada: Dependiendo de la cantidad de datos que se hayan utilizado para calcular el valor muestral, este se acercar ms o menos al verdadero parmetro poblacional.
- Nivel de confianza: Nos va a informar en qu porcentaje de casos nuestra estimacin acierta. Los niveles habituales son el 95% y el 99%.
- Margen de error de nuestra estimacin: Este se denomina como alfa y nos informa de la probabilidad que existe de que el valor poblacional est fuera de nuestro intervalo.
- Lo estimado en la muestra (media, varianza, diferencia de medias): De esto va a depender el estadstico pivote para el clculo del intervalo.

# Error estndar

El **error estndar (EE o SE *Standard Error*)** es un concepto importante en los intervalos de confianza, no tiene aplicacin para expresar la variabilidad de los individuos de una poblacin, es slo un *indicador de la variabilidad* de las medias calculadas en muchas posibles muestras que se tomen de una poblacin, todas ellas de tamao n.

El error estndar es a la muestra, lo que la desviacin estndar es al individuo.As pues, el error estndar de la media mide nuestro grado de incertidumbre respecto a la capacidad de la media muestral para estimar la media poblacional.

El error estndar de la media (llamado en ingls **"standard error of the mean" (SEM))** cuantifica las oscilaciones de la media muestral (media obtenida en los datos) alrededor de la media poblacional (verdadero valor de la media): se estima generalmente dividiendo la desviacin estndar de la poblacin entre la raz cuadrada del tamao de la muestra (suponiendo independencia estadstica de los valores en la muestra):

# Ejercicio 1: intervalo de confianza para la media

```python
import pandas as pd
import numpy as np
import scipy.stats as st
```

```python
# Distribucion Normal loc=12 (media 12) y scale=10 (es decir varianza 10)
X = st.norm(loc=12.0, scale=10.0) 
dataframeejer1 = pd.DataFrame(X.rvs(size=10000).reshape(100,100))
```

```python
def intervalo_confianza_media(data, sigma=1.0, alpha=0.05):
    media = data.mean()
    h = st.norm.ppf(1.0-(alpha/2.)) * (sigma / np.sqrt(data.size))
    #devuelve media , media-h limite inferior, media+h limite superior
    return media, media-h, media+h
```

```python
a = dataframeejer1.iloc[:,0]

#a = samples.iloc[:,1] cambia la muestra cambia el intervalo de confianza
intervalo_confianza_media(a, sigma=10.0)
```

## Ejercicio 2
Crear una funcin que calcule el intervalo de confianza para la varianza poblacional

```python
def var_confidence_interval_sinSigma(data,alpha=0.05):
    v = data.var()
    l_inf = (data.size-1.0)*data.var()/st.chi2.ppf(1.-alpha/2., data.size-1.0)
    l_sup = (data.size-1.0)*data.var()/st.chi2.ppf(alpha/2., data.size-1.0)
    #v varianza ,l_inf limite inferior ,l_sup limite superior 
    return v, l_inf, l_sup
```

```python
X = st.norm(loc=12.0, scale=10.0)
samples = pd.DataFrame(X.rvs(size=10000).reshape(100,100))
```

```python
a = samples.loc[:,0]

var_confidence_interval_sinSigma(a, alpha=0.10) 
```

## Ejercicio 3
Intervalo de confianza para una proporcin poblacional aproximado por la normal.
Funcion con n individuos con una caracteristica, a los que he preguntado y alpha
def intervaloConfianzaAproximado(m,p,alpha=0.05)

```python
def intervaloConfianzaAproximado(m,p,alpha=0.05): 
    l_inf=m/p-st.norm.ppf(1-alpha/2)*np.sqrt((m/p)*(1-m/p)/p) - 1/(2.*p)    
    l_sup=m/p+st.norm.ppf(1-alpha/2)*np.sqrt((m/p)*(1-m/p)/p) - 1/(2.*p)    
    return m/p,l_inf,l_sup # valor limites inf y sup y valor 
```

```python
intervaloConfianzaAproximado(m=20.0,p=100.0,alpha=0.01)
```

```python
def intervaloConfianza(n,p,alpha): 
    
    limite_inf=(p/n)-st.norm.ppf(1-alpha/2)*np.sqrt((p/n)*(1-p/n)/n)    
    limite_sup=(p/n)+st.norm.ppf(1-alpha/2)*np.sqrt((p/n)*(1-p/n)/n)    
    
    return p/n,limite_inf,limite_sup # valor limites inf y sup y valor 
```

## Ejercicio 4
Intervalo de confianza para el cociente de varianzas 

```python
def IC_cociente_varianzas (data1,data2,alpha=0.05):
    var1=data1.var()
    var2=data2.var()
    m=data1.size
    n=data2.size
    l_inf=st.f.ppf(alpha/2.,m-1,n-1)*var1/var2
    l_sup=st.f.ppf(1.0-alpha/2.,m-1,n-1)*var1/var2
    return var1/var2,l_inf,l_sup
```

## Solucin

El intervalo ser de la forma:



Es decir, Sabemos que de cada cien muestras que consideremos, el valor autntico de la  estar, como mnimo, en
noventa y cinco de los intervalos correspondientes. Por eso, si el valor que queremos contrastar a la hiptesis nula cae dentro del intervalo, aceptaremos la hiptesis nula.

```python
import numpy as np
import scipy.stats as st

def intervalo_confianza_pedido(tamao_muestra=1.0,media=1.0,sigma=1.0, alpha=0.05):
    h = st.norm.ppf(1.0-(alpha/2.)) * (sigma / np.sqrt(tamao_muestra))
    #devuelve media , media-h limite inferior, media+h limite superior
    return media-h, media+h

media=163.8
n=42
desv_tipica=3.5

intervalo_confianza_pedido(n, media, desv_tipica)
```
