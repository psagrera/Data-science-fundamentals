
# Sobre intervalos de Confianza

# Estimador: *tipos*
Una estimación consiste en apostar por un valor para un parámetro poblacional, habitualmente calculando también un rango de valores entre los que puede situarse dicho parámetro poblacional. Por tanto, un intervalo de confianza es simplemente un rango de valores en el que confiamos que se contenga el parámetro poblacional (Altman, 1991).
>Puede decirse que el análisis estadístico inferencial conduce a dos clases de resultados: valores “p” e intervalos de confianza.

Las estimaciones pueden ser de dos tipos:
>**Estimación puntual**: obtendremos un punto, un valor, como estimación del parámetro desconocido, quedando sin especificar cómo de buena es la aproximación.
>> Si a partir de una muestra se afirma que el 70% de una población es fumadora se está haciendo una estimación puntual de la proporción de fumadores.

>**Estimación por intervalo**: proporciona un rango de posibles valores, dentro del cual estimamos (bajo cierta probabilidad) que estará el parámetro desconocido.
>>Si a partir de una muestra se afirma que el porcentaje de la población que es fumadora se sitúa entre un 23 y 52% se está haciendo una estimación por intervalo de la proporción poblacional de fumadores



En general, se puede afirmar que un estimador puede ser:
+ **Insesgado**: cuando el estimador muestral coincide con el poblacional, sin ninguna desviación sistemática. Es decir, un estimador es insesgado si su valor medio sobre todas las posibles muestras de tamaño n coincide con el parámetro poblacional.
+ **Sesgado**: cuando el estimador muestral no se corresponde con el poblacional, debido a un error sistemático.
+ **Eficiente**: es aquel estimador muestral de igual valor a otros, que la varianza muestral es mínima. De esta forma, se tendrá una mayor confianza en que la estimación resultante de la muestra finalmente seleccionada esté próxima al parámetro poblacional.
+ **Suficiente**: aquel estimador que utiliza toda la información contenida en la muestra.
>>Un buen estimador es aquel que es insesgado, eficiente y suficiente
(Rothman y Greenland, 1998)



# ¿Qué es un intervalo de confianza?

Las estimaciones puntuales obtenidas a partir de una muestra *diferirán del parámetro poblacional* y, en consecuencia, quedará un *margen de incertidumbre* que se expresa en términos del error estándar del
estimador. 

Por consiguiente, podemos querer disponer de una medida del parámetro poblacional que incorpore tanto la estimación puntual como su error estándar. Esta medida es el **intervalo de confianza (I.C.)**, *que facilita un rango de valores dentro del cual se encontrará el verdadero valor del parámetro poblacional con un cierto grado de confianza.*

# Factores de los que depende un intervalo de confianza
El cálculo de un intervalo de confianza depende principalmente de los siguientes factores:
- Tamaño de la muestra seleccionada: Dependiendo de la cantidad de datos que se hayan utilizado para calcular el valor muestral, este se acercará más o menos al verdadero parámetro poblacional.
- Nivel de confianza: Nos va a informar en qué porcentaje de casos nuestra estimación acierta. Los niveles habituales son el 95% y el 99%.
- Margen de error de nuestra estimación: Este se denomina como alfa y nos informa de la probabilidad que existe de que el valor poblacional esté fuera de nuestro intervalo.
- Lo estimado en la muestra (media, varianza, diferencia de medias…): De esto va a depender el estadístico pivote para el cálculo del intervalo.

# Error estándar

El **error estándar (EE o SE *Standard Error*)** es un concepto importante en los intervalos de confianza, no tiene aplicación para expresar la variabilidad de los individuos de una población, es sólo un *indicador de la variabilidad* de las medias calculadas en muchas posibles muestras que se tomen de una población, todas ellas de tamaño n.

El error estándar es a la muestra, lo que la desviación estándar es al individuo.Así pues, el error estándar de la media mide nuestro grado de incertidumbre respecto a la capacidad de la media muestral para estimar la media poblacional.

El error estándar de la media (llamado en inglés **"standard error of the mean" (SEM))** cuantifica las oscilaciones de la media muestral (media obtenida en los datos) alrededor de la media poblacional (verdadero valor de la media): se estima generalmente dividiendo la desviación estándar de la población entre la raíz cuadrada del tamaño de la muestra (suponiendo independencia estadística de los valores en la muestra):


si conocemos la desviación estándar de la población, y:

si usamos la desviación estándar de la muestra.
donde:
+ σ:la desviación estándar de la población
+ s: desviación estándar de la muestra
+ N: número de observaciones de la muestra

Esta medida se conoce como el error estándar y tiende a disminuir cuando aumenta el tamaño de la(s) muestra(s).

```python
def intervalo_confianza(data):
    media=data.mean()
    desviacion=data.std()
    SE=desviacion/np.sqrt(data.iloc[:,0].size)
    inferior=media-SE
    superior=media+SE
    return desviacion, media, SE, inferior, superior
```

# Ejercicio 1: intervalo de confianza para la media

```python
import pandas as pd
import numpy as np
import scipy.stats as st
```

```python
# Distribucion Normal loc=12 (media 12) y scale=10 (es decir varianza 10)
X = st.norm(loc=12.0, scale=10.0) 
dataframeejer1 = pd.DataFrame(X.rvs(size=10000).reshape(100,100))
```

```python
def intervalo_confianza_media(data, sigma=1.0, alpha=0.05):
    media = data.mean()
    h = st.norm.ppf(1.0-(alpha/2.)) * (sigma / np.sqrt(data.size))
    #devuelve media , media-h limite inferior, media+h limite superior
    return media, media-h, media+h
```

```python
a = dataframeejer1.iloc[:,0]

#a = samples.iloc[:,1] cambia la muestra cambia el intervalo de confianza
intervalo_confianza_media(a, sigma=10.0)
```

## Ejercicio 2
Crear una función que calcule el intervalo de confianza para la varianza poblacional

```python
def var_confidence_interval_sinSigma(data,alpha=0.05):
    v = data.var()
    l_inf = (data.size-1.0)*data.var()/st.chi2.ppf(1.-alpha/2., data.size-1.0)
    l_sup = (data.size-1.0)*data.var()/st.chi2.ppf(alpha/2., data.size-1.0)
    #v varianza ,l_inf limite inferior ,l_sup limite superior 
    return v, l_inf, l_sup
```

```python
X = st.norm(loc=12.0, scale=10.0)
samples = pd.DataFrame(X.rvs(size=10000).reshape(100,100))
```

```python
a = samples.loc[:,0]

var_confidence_interval_sinSigma(a, alpha=0.10) 
```

## Ejercicio 3
Intervalo de confianza para una proporción poblacional aproximado por la normal.
Funcion con n individuos con una caracteristica, a los que he preguntado y alpha
def intervaloConfianzaAproximado(m,p,alpha=0.05)

```python
def intervaloConfianzaAproximado(m,p,alpha=0.05): 
    l_inf=m/p-st.norm.ppf(1-alpha/2)*np.sqrt((m/p)*(1-m/p)/p) - 1/(2.*p)    
    l_sup=m/p+st.norm.ppf(1-alpha/2)*np.sqrt((m/p)*(1-m/p)/p) - 1/(2.*p)    
    return m/p,l_inf,l_sup # valor limites inf y sup y valor 
```

```python
intervaloConfianzaAproximado(m=20.0,p=100.0,alpha=0.01)
```

```python
def intervaloConfianza(n,p,alpha): 
    
    limite_inf=(p/n)-st.norm.ppf(1-alpha/2)*np.sqrt((p/n)*(1-p/n)/n)    
    limite_sup=(p/n)+st.norm.ppf(1-alpha/2)*np.sqrt((p/n)*(1-p/n)/n)    
    
    return p/n,limite_inf,limite_sup # valor limites inf y sup y valor 
```

```python
intervaloConfianza(40.0,24.0,0.05)
```

## Ejercicio 4
Intervalo de confianza para el cociente de varianzas 

```python
def IC_cociente_varianzas (data1,data2,alpha=0.05):
    var1=data1.var()
    var2=data2.var()
    m=data1.size
    n=data2.size
    l_inf=st.f.ppf(alpha/2.,m-1,n-1)*var1/var2
    l_sup=st.f.ppf(1.0-alpha/2.,m-1,n-1)*var1/var2
    return var1/var2,l_inf,l_sup
```

# Ejercicio 5
El peso en gramos de las manzanas de una explotación agraria sigue una distribución normal de desviación típica 3,5. Un estudio nos dice que el peso medio es de 155gr y **queremos comprobar si eso es cierto**.

Para hacerlo, hemos cogido una muestra aleatoria de 42 manzanas y hemos obtenido una media de 163,8.

Encontrar un intervalo de confianza del 95% para la media de la distribución.

## Solución
Queremos contrastar  $H_0$: μ = 155 contra &H_1&: μ ≠ 155

El intervalo será de la forma:



Es decir, Sabemos que de cada cien muestras que consideremos, el valor auténtico de la μ estará, como mínimo, en
noventa y cinco de los intervalos correspondientes. Por eso, si el valor que queremos contrastar a la hipótesis nula cae dentro del intervalo, aceptaremos la hipótesis nula.

```python
import numpy as np
import scipy.stats as st

def intervalo_confianza_pedido(tamaño_muestra=1.0,media=1.0,sigma=1.0, alpha=0.05):
    h = st.norm.ppf(1.0-(alpha/2.)) * (sigma / np.sqrt(tamaño_muestra))
    #devuelve media , media-h limite inferior, media+h limite superior
    return media-h, media+h

media=163.8
n=42
desv_tipica=3.5

intervalo_confianza_pedido(n, media, desv_tipica)
```

En este caso, puesto que 155 no peretenece al intervalo dado (162,74, 164,85), debemos **rechazar la hipótesis
nula**.
