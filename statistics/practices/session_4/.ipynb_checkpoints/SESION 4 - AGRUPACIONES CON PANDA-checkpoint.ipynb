### Agregar datos por grupo utilizando el método *groupby* de Pandas Groupby

La mayoría de las veces queremos tener nuestras estadísticas resumidas en la misma tabla. 
Podemos calcular el salario medio y mediano, por grupos, utilizando el método *agg* . 
Ejemplo, agregemos el salario mínimo y máximo por grupo (rango):

```python
dataframe_rank['salary'].agg(['mean', 'median', 
                                  'std', 'min', 'max']).reset_index()
```

Una cosa muy interesante con el método *agg* de Pandas, es que podemos *escribir funciones personalizadas* y pasarlas. 
Supongamos que queremos, en lugar de tener una columna para salario mínimo y una columna para salario máximo, tener una columna con el rango de salario:

```python
def salary_range(df):
    mini = df.min()
    maxi = df.max()
    rang = '%s - %s' % (mini, maxi)
    
    return rang

df_descriptive = dataframe_rank['salary'].agg(['mean', 'median', 'std', salary_range]).reset_index()
```

Aquí, sin embargo, la salida tendrá el nombre de los métodos / funciones utilizados. Es decir, tendremos una columna llamada *'salary_range'* y vamos a cambiar el nombre de esta columna::

```python
# Renaming Pandas Dataframe Columns
df_descriptive.rename(columns={'rank':'Rank', 'mean':'Mean', 'median':'Median', 
                               'std':'Standard Deviation', 'salary_range':'Range'})
```

Además, es posible utilizar métodos de otros paquetes de Python como SciPy y NumPy.

## Pandas Groupby Count Multiple Groups

En el siguiente grupo, por ejemplo, vamos a calcular el número de observaciones en tres grupos (es decir, "n"). Tenemos que comenzar agrupando por "rank", "discipline" y "sex" usando groupby. 
Al igual que en el ejemplo anterior (grupo por una columna), utilizamos el tamaño del método para calcular el  *ny reset_index*, con el nombre del parámetro  = "n" , para llevar la serie a un marco de datos:

```python
df_3grps = dataframe.groupby(['rank', 'discipline', 'sex'])
df_n_per_group = df_3grps.size().reset_index(name='n')
```

Calculamos el porcentaje por grupo:

```python
perc = dataframe.groupby(['rank', 'discipline', 'sex'])['salary'].size()

# Give the percentage on the level of Rank:
percbyrank = perc.groupby(level=0).apply(lambda x: 100 * x / float(x.sum()))

print(percbyrank)
print('Total percentage in group AssocProf. ',
      percbyrank.reset_index().query('rank == "AssocProf"')['salary'].sum())
```

Cálculo del número de mujeres y hombres:

```python
n = perc.reset_index()['salary'].sum()
totalperc =  perc.groupby(level=0).apply(lambda x:100*x/n).reset_index(name='% of total n')
totalperc.reset_index()
```

## Usar groupby agg con varios grupos

Usaremos agg junto con múltiples grupos.

Es decir, vamos a calcular la media, la mediana y la desviación estándar utilizando el método agg. En este grupo, por ejemplo, también estamos agregando las estadísticas de resumen (es decir, "media", "mediana" y "estándar") a cada columna. 

De lo contrario, obtendremos un resultado indexado de varios niveles como la imagen a continuación:


```python
df_stats = dataframe.groupby(['rank', 'discipline', 'sex']).agg(['mean', 'median', 'std'])
df_stats.columns = ["_".join(x) for x in df_stats.columns.ravel()]

df_stats.iloc[:,0:6].reset_index()
```

Hemos usado *df.iloc* en la última línea. Lo que hicimos fue tomar las primeras seis columnas, usando iloc. Después de eso, reiniciamos el índice y obtuvimos ocho columnas en total.
