Una variable aleatoria es toda funcin cuyos valores numricos,

## Distribucin Bernoulli: $\mathcal{B}(p)$
<hr>

A los experimentos aleatorios en los que solo se pueden dar dos resultados posibles: xito y fracaso.

__[EJEMPLO]__: Lanzamiento de una moneda trucada: la probabilidad de salir cara (xito) es 0.8

$$X=\begin{cases}
0 & \mbox{sale cruz}\\
1 & \mbox{sale cara}
\end{cases}
$$

Entonces $X \sim \mathcal{B}(0.8)$

La __funcin masa de probabilidad__ (_pmf_ en sus siglas en ingls) es 

$$p_1=P(X=x_1)=P(X=0)=1-p, \quad p_2=P(X=x_2)=P(X=1)= p$$

Para el ejemplo anterior, 

|Valor X          |0  |1  |
|-----------------|---|---|
|$p_i$       |0.2|0.8|


```python
%pylab inline
%config InlineBackend.figure_format = 'retina'
pylab.rcParams['figure.figsize'] = (10, 6)
plt.style.use('ggplot')

import scipy.stats as st
import pandas as pd
import numpy as np 

fig, ax = plt.subplots(1, 1)
# Probabilidad
p = 0.8

x = np.array([0.0, 1.0])
ax.plot(x, st.bernoulli.pmf(x, p), 'bo', ms=8, label='Bernoulli pmf')

# Modelo de Bernoulli
ax.vlines(x, 0, st.bernoulli.pmf(x, p), colors='b', lw=5, alpha=0.5);
```

## Distribucin Binomial: $\mathcal{Bi}(n,p)$
<hr/>

Si un experimento de Bernoulli se realiza consecutivamente $n$ veces, de forma independiente, y siempre en las mismas condiciones que la primera vez, entonces podemos calcualr el **nmero total de xitos** conseguidos en las $n$ repeticiones. 

En ese caso, $X$, el nmero de xitos en $n$ intentos, con probabilidad de xito $p$ sigue una distribucin Binomial: $\mathcal{Bi}(n,p)$. La __funcin masa de probabilidad__ (pmf)

$$P(X=k) = {n \choose k} p^k \cdot (1-p)^{n-k}$$

__[EJEMPLO]__: Nmero de choches vendidos, con una probabilidad de venta del coche de 0.31. En el concesionario hay 10 coches.
Entonces $$X\sim \mathcal{Bi}(10, 0.31)$$


```python
# 10 coches
n = 10 

# Probabilidad
p = 0.31 

x = np.arange(11) # del 1 al 10
x
```

```python
st.binom.pmf(x, n, p)
```

```python
fig, ax = plt.subplots(1,2,figsize=(15,8))

x = np.arange(st.binom.ppf(0.01, n, p),st.binom.ppf(0.99, n, p))
ax[0].plot(x, st.binom.pmf(x, n, p), 'bo', ms=8, label='binom pmf')
ax[0].vlines(x, 0, st.binom.pmf(x, n, p), colors='b', lw=5, alpha=0.5);


xx = np.arange(15)
ax[1].plot(xx, st.binom.cdf(xx, n, p), 'bo'); # Funcin de distribucin (cdf en sus siglas en ingls)
```

## Distribucin Poisson: $\mathcal{Po}(\lambda)$
<hr>

Describe el nmero de veces que ocurre un suceso por unidad de tiempo o espacio:

* Avisos de Whatsapp que recibimos en nuestro mvil cada da.
* Pacientes que asisten a urgencias en un Centro de Salud al ao.

La distribucin queda definida por el parmetro $\lambda$ que representa el nmero medio de ocurrencias del suceso. 

Por ejemplo, Si $X$ representa el nmero de avisos de Whatsapp que recibimos en nuestro mvil cada da, y modelamos $X$ como

$$ X \sim \mathcal{Po} (54)$$ 

significa que recibimos como media al da 54 avisos al mvil. 

La funcin masa de probabilidad es 

$$P(X=k) = \frac{e^{-\lambda} \cdot \lambda^k}{k!}$$

```python
fig, ax = plt.subplots(1, 1)

#l es 
l = 54

x = np.arange(st.poisson.ppf(0.01, l),st.poisson.ppf(0.99,l))
ax.plot(x, st.poisson.pmf(x, l), 'bo', ms=8, label='Poisson pmf')
ax.vlines(x, 0, st.poisson.pmf(x, l), colors='b', lw=5, alpha=0.5);
```

## Distribucin normal
<hr>

Tiene como __funcin de densidad__ (no confundir con funcin masa de probabilidad):

$$f(x|\mu, \sigma) = \frac{1}{\sqrt{2\pi \sigma^2}} \cdot e^{-\frac{(x-\mu)^2}{2\sigma^2}}$$

Esta funcin permite calcular la probabilidad de intervalos en los que est definida la variable aleatoria. Por ejemplo, la variable $X$ podra medir los ingresos en miles de euros de varios departamentos cada da del mes, entonces

$$X \sim \mathcal{N} (5,2)$$

y por tanto su funcin de densidad sera

$$f(x|5,2) = \frac{1}{\sqrt{8\pi}} \cdot e^{-\frac{(x-5)^2}{8}}$$

```python
fig, ax = plt.subplots(1,1)

# Media: loc= 5  
# Desviacin Tpica: scale 2 
data = pd.DataFrame(st.norm(loc=5, scale=2).rvs(100))
data.plot(kind='hist', normed = 'True', ax = ax)

x = np.linspace(-5,15,1000)
y = st.norm(loc=5, scale=2).pdf(x)
ax.plot(x,y,color='orange');
ax.legend_.remove()
```

```python
fig, ax = plt.subplots(1,1)

x = np.linspace(-15,20,1000)

y1 = st.norm(loc=5, scale=2).pdf(x)
ax.plot(x,y1,color='orange');

y2 = st.norm(loc=5, scale=5).pdf(x)
ax.plot(x,y2,color='red');

y3 = st.norm(loc=5, scale=10).pdf(x)
ax.plot(x,y3,color='yellow');

y4 = st.norm(loc=0, scale=2).pdf(x)
ax.plot(x,y4,color='blue');

y5 = st.norm(loc=0, scale=1).pdf(x)
ax.plot(x,y5,color='black');#normal
```

## Distribucin Chi-cuadrado: $\chi^2(m)$
<hr>

Se define como la suma de distribuciones normales estandarizadas independientes al cuadrado, es decir,

$$Y = \sum_{i=1}^m Z_i^2, \quad \mbox{siendo} \quad Z_i \sim \mathcal{N}(0,1)$$

El nico parmetro en esta distribucin es saber cuntas normales estandarizadas tenemos que sumar $(m)$, tambin llamado _grados de libertad_ (df en sus siglas en ingls). La funcin de densidad es 

$$ f(x| m) = \frac{1}{2\cdot \Gamma(m/2)} \cdot \left(\frac{x}{2}\right)^{m/2-1} \cdot e^{-x/2} $$

```python
fig, ax = plt.subplots(1, 1)

# Grados de Libertad
df = 5

x = np.linspace(st.chi2.ppf(0.01, df), st.chi2.ppf(0.99, df), 100)
ax.plot(x, st.chi2.pdf(x, df), 'r-', lw=5, alpha=0.6, label='chi2 pdf');
```

## Distrubucin t-Student
<hr>

Consideremos dos variables aleatorias $X \sim  \mathcal{N}(0,1)$ e $Y \sim \chi^2(m)$, entonces la variable

$$ T = \frac{X}{\sqrt{\frac{Y}{m}}} $$

se distribuye como una distribucin t-Student con $m$ grados de libertad, $t(m)$, con funcin de densidad

$$f(x|m) = \frac{\Gamma((m+1)/2)}{\sqrt{\pi \cdot m} \cdot \Gamma(df/2) \cdot (1+x^2/m)^{(m+1)/2}} $$

```python
fig, ax = plt.subplots(1,1)

data = pd.DataFrame(st.t(3).rvs(100))
data.plot(kind='hist', normed = 'True', ax = ax)

x = np.linspace(-15,15,1000)
y = st.t(3).pdf(x)
ax.plot(x,y,color='orange');
ax.legend_.remove()
```

```python
# Existe una t-Student para cada uno de los grados de libertad
fig, ax = plt.subplots(1,1)

x = np.linspace(-15,15,1000)

y0 = st.t(1).pdf(x)
ax.plot(x,y0,color='red');

y1 = st.t(3).pdf(x)
ax.plot(x,y1,color='orange');

y2 = st.t(10).pdf(x)
ax.plot(x,y2,color='blue');

y3 = st.t(20).pdf(x)
ax.plot(x,y2,color='yellow');

y4 = st.t(100).pdf(x)
ax.plot(x,y4,color='black');
```

```python
fig, ax = plt.subplots(1,1)

x = np.linspace(-15,15,1000)

y1 = st.t(3,loc=5).pdf(x)
ax.plot(x,y1,color='orange');

y2 = st.t(3,loc=5, scale=5).pdf(x)
ax.plot(x,y2,color='red');
```
