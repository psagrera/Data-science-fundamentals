# Solución

```python
import pandas as pd
collection = pd.read_csv('collectionsEE2013.csv')
collection.head()
collection.info()
```

```python
collection
```

```python
print("Media: \n",collection.mean(axis=0))
print("Mediana: \n",collection.median(axis=0))
print("Maximo: \n",collection.max(axis=0))
print("Minimp: \n",collection.min(axis=0))
```

```python
print("Media: \n",collection.mean(axis=1))
print("Mediana: \n",collection.median(axis=1))
print("Maximo: \n",collection.max(axis=1))
print("Minimo: \n",collection.min(axis=1))
```

```python
df_media = collection.iloc[:,:] - collection.mean(axis=0)
print(df_media)
print()
df_max = collection.iloc[:,:] - collection.max(axis=0)
print(df_max)
print()
df_min = collection.iloc[:,:] - collection.min(axis=0)
print(df_min)
print()
df_mediana = collection.iloc[:,:] - collection.median(axis=0)
print(df_mediana)
```

```python
df_media = pd.DataFrame()
df_max = pd.DataFrame()
df_min = pd.DataFrame()
df_mediana = pd.DataFrame()
for etiqueta,fila in collection.iterrows():
    #print (etiqueta,fila)
    media=fila-fila.mean()
    df_media = df_media.append(media)
    maximo = fila - fila.max()
    df_max = df_max.append(maximo)
    minimo = fila - fila.min()
    df_min = df_min.append(minimo)
    mediana = fila - fila.median()
    df_mediana = df_mediana.append(mediana)
print() 
print (df_media)
print()
print(df_max)
print()
print (df_min)
print()
print(mediana)
```

```python
writertexto = pd.ExcelWriter('diferenciasPorTexto.xlsx')
df_media.to_excel(writertexto,"mediafila")
df_mediana.to_excel(writertexto,"medianafila")
df_max.to_excel(writertexto,"maxfila")
df_min.to_excel(writertexto,"minfila")
writertexto.save()
```
