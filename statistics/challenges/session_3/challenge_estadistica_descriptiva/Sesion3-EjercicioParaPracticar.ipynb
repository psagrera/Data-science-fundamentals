# Ejercicio

Data la fuente de datos drinks.csv, se pide: 
* Calcular el histograma para la variable *total_litres_of_pure_alcohol*. 
* Calcular la media, mediana y moda para la variable *total_litres_of_pure_alcohol*. 
* Calcular entre qué dos valores está el 50% central de la muestra para la variable *total_litres_of_pure_alcohol*.
* Considerando el 20% de las mayores cantidades de litros totales de alcohol puro consumido, ¿cuál es la menor de ellas?

```python
import pandas as pd

drinks = pd.read_csv('drinks.csv')
drinks.head()
drinks.info()
#drinks.total_litres_of_pure_alcohol.value_counts()
```

```python
drinks.total_litres_of_pure_alcohol.value_counts(normalize=True) #frecuencia relativa de una variable cuantitativa
drinks.total_litres_of_pure_alcohol.plot(kind='hist', normed=True); # Mejor densidades si hay intervalos con diferente longitud!!!
```

```python
drinks.groupby('country').mean().plot(kind='bar')
```

```python
drinks.total_litres_of_pure_alcohol.mean()
drinks.total_litres_of_pure_alcohol.median()
drinks.total_litres_of_pure_alcohol.mode()
```

```python
drinks.total_litres_of_pure_alcohol.quantile([.25, .75]) #parte central el 50% es entre el 25 y 75
```

```python
drinks.total_litres_of_pure_alcohol.quantile([.80])
```
