## Conjunto de entrenamiento y test

Crea un conjunto de entrenamiento y test con una proporción del 30%.

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(dataset_X, dataset_Y, test_size=.3, random_state=42) 
```

```python
# Funcion para obetener algunas metricas relativas al funcionamiento del modelo
from sklearn import metrics

def print_evaluate(true,predicted):
    mae = metrics.mean_absolute_error(true,predicted)
    mse = metrics.mean_squared_error(true, predicted)
    rmse = np.sqrt(metrics.mean_squared_error(true,predicted))
    r2_square = metrics.r2_score(true, predicted)
    print('\nMAE:', mae)
    print('\nMSE:', mse)
    print('\nRMSE:', rmse)
    print('\nR2 Square', r2_square)
```

## Entrenamiento y clasificación

Utiliza un RandomForest como algoritmo de clasificación y entrena el modelo.

```python

from sklearn.ensemble import RandomForestClassifier
random_forest = RandomForestClassifier()
random_forest.fit(X_train, y_train)
print_evaluate(y_test,random_forest.predict(X_test))
```

```python
random_forest.score(X_test, y_test)
```

```python
y_predicted = random_forest.predict(X_test)
print("accuracy: {}".format(accuracy_score(y_test,y_predicted))) 
print("precision_score: {}".format(precision_score(y_test,y_predicted,average='weighted'))) # Este parametro es requerido para multiclase / multilabel targets
print("recall_score: {}".format(recall_score(y_test,y_predicted,average='weighted')))
print("F1: {}".format(f1_score(y_test,y_predicted,average='weighted'))) # Metrica que calcula la media geometrica entre el precision y el recall
```

## Evalua el modelo

Calcula score, ROC y matriz de confusión

```python
import scikitplot as skplt
import matplotlib.pyplot as plt

skplt.metrics.plot_roc(y_test, random_forest.predict_proba(X_test))
plt.show()
```

```python
import scikitplot as skplt
y_predicted = random_forest.predict(X_test)
skplt.metrics.plot_confusion_matrix(y_test, y_predicted)
```
