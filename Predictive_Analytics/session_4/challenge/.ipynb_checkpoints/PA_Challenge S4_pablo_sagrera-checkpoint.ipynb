# Challenge S3: Prophet 
# Personas que ven un escaparate de una tienda

Una cadena de retail está interesada en realizar una previsión de las personas que ven el escaparate de su tienda ya que es una variable que afecta directamente a las ventas de la tienda. Durante dos años ha generado un histórico diario de personas que se han parado a ver el escaparate de una tienda céntrica de Madrid, colocando un dispositivo de IoT que detecta el número de personas que se paran en una tienda y deciden que sólo si la persona permanece más de 3 segundos parado en frente del escaparate se contará como persona que ha visto el escaparate.

Nos piden realizar una previsión para el próximo año a los datos históricos que tenemos.

Nuestro objetivo es:
1. Cargar las ventas en un DataFrame (tenemos un csv con los datos de partida: personas_escaparate.csv)
2. Preparar el DataFrame para cargar en un modelo que asumiremos como ADITIVO en Prophet.
3. Ajustar el modelo.
4. Analizar las previsiones y errores.
5. Mejorar nuestro modelo de previsión incluyendo algún evento especial.

```python
import pandas as pd
import numpy as np
from fbprophet import Prophet
import matplotlib.pyplot as plt
%matplotlib inline
import fbprophet

print(fbprophet.__version__)
```

## 1. Cargar las ventas en un DataFrame

```python
df_datos = pd.read_csv('personas_escaparate.csv')
df_datos
df_datos['fecha']=pd.to_datetime(df_datos['fecha'],dayfirst=True)
```

## 2. Preparar el DataFrame para cargar en un modelo Prophet.

```python
df_datos= df_datos.rename(columns={'fecha': 'ds', 'valor': 'y'})
df_datos


ax = df_datos.set_index('ds').plot(figsize=(12, 8))
ax.set_ylabel('Personas')
ax.set_xlabel('Fecha')
plt.show()
# Posibles picos
filtro  = df_datos['y'] > 5000
df_datos[filtro]
```

## 3. Ajustar el modelo

```python
modelo = Prophet()
modelo.fit(df_datos)
```

## 4. Analizar las previsiones y errores.

```python
futuro = modelo.make_future_dataframe(periods=365)
```

Visualización de las componentes de la serie

```python
prevision = modelo.predict(futuro)
prevision.tail()
```

```python
modelo.plot(prevision);
```

```python
modelo.plot_components(prevision);
```

```python
eventos_marzo = pd.DataFrame({
  'holiday': 'easter',
  'ds': pd.to_datetime(['2015-03-29','2015-03-30','2015-03-31','2015-04-01','2015-04-02','2015-04-03','2015-04-04',
                        '2016-03-20','2016-03-21','2016-03-22','2016-03-23','2016-03-24','2016-03-25','2016-03-26',
                        '2017-04-09','2017-04-10','2017-04-11','2017-04-12','2017-04-13','2017-04-14','2017-04-15']),
  'lower_window': 0,
  'upper_window': 1, # Como ventanas de efecto (i.e van a tener efecto ese dia y el siguiente)
})
eventos_marzo
```

## 5. Mejorar nuestro modelo de previsión incluyendo algún evento especial.

```python
eventos_marzo = eventos_marzo.reset_index()  # THE FIX
modelo2 = Prophet(holidays=eventos_marzo)
prevision2 = modelo2.fit(df_datos).predict(futuro)
```

Comparativa de ambos modelos

```python
modelo.plot(prevision);
```

```python
modelo.plot(prevision2);
```

```python
modelo.plot_components(prevision2);
```

```python
# R2 para el modelo sin eventos especiales
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error

metric_df = prevision.set_index('ds')[['yhat']].join(df_datos.set_index('ds').y).reset_index()
metric_df.dropna(inplace=True)

r2_score(metric_df.y, metric_df.yhat)
```

```python
# R2 para el modelo con eventos especiales
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error

metric_df = prevision2.set_index('ds')[['yhat']].join(df_datos.set_index('ds').y).reset_index()
metric_df.dropna(inplace=True)

r2_score(metric_df.y, metric_df.yhat)
```
